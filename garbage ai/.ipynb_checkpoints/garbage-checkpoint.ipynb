```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import time
import warnings
import pickle
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from IPython.display import Image, display
import os

warnings.filterwarnings("ignore", category=UserWarning, message="Geometry is in a geographic CRS.*")
print(f"Using osmnx version: {ox.__version__}")

localities = [
    "Lodhi Colony, Delhi, India",
    "Cuffe Parade, Mumbai, India",
    "Cubbon Park, Bengaluru, India",
    "Lalbagh, Bengaluru, India",
    "Besant Nagar, Chennai, India",
    "Adyar, Chennai, India",
    "Gachibowli, Hyderabad, India",
    "Banjara Hills, Hyderabad, India",
    "Alipore, Kolkata, India",
    "New Town, Kolkata, India",
    "Sector 62, Noida, India",
    "Whitefield, Bengaluru, India",
    "Koregaon Park, Pune, India",
    "Rajendra Nagar, Indore, India",
    "Maninagar, Ahmedabad, India",
    "Vyttila, Kochi, India",
    "Panaji, Goa, India",
    "Aundh, Pune, India",
    "Anna Nagar, Chennai, India",
    "Shyambazar, Kolkata, India",
    "Jubilee Hills, Hyderabad, India",
    "Powai, Mumbai, India",
    "Salt Lake City, Kolkata, India",
    "DLF Phase 3, Gurugram, India",
    "Jayanagar, Bengaluru, India",
    "Porvorim, Goa, India"
]

fallbacks = {
    "Gachibowli, Hyderabad, India": "Hyderabad, India",
    "Banjara Hills, Hyderabad, India": "Hyderabad, India",
    "New Town, Kolkata, India": "Kolkata, India",
    "Besant Nagar, Chennai, India": "Chennai, India",
    "Powai, Mumbai, India": "Mumbai, India",
    "Salt Lake City, Kolkata, India": "Kolkata, India",
    "Alipore, Kolkata, India": "Kolkata, India",
    "Sector 62, Noida, India": "Noida, India",
    "Whitefield, Bengaluru, India": "Bengaluru, India",
    "Rajendra Nagar, Indore, India": "Indore, India",
    "Maninagar, Ahmedabad, India": "Ahmedabad, India",
    "Vyttila, Kochi, India": "Kochi, India",
    "Panaji, Goa, India": "Goa, India",
    "Aundh, Pune, India": "Pune, India",
    "Anna Nagar, Chennai, India": "Chennai, India",
    "Shyambazar, Kolkata, India": "Kolkata, India",
    "Jubilee Hills, Hyderabad, India": "Hyderabad, India",
    "DLF Phase 3, Gurugram, India": "Gurugram, India",
    "Jayanagar, Bengaluru, India": "Bengaluru, India",
    "Porvorim, Goa, India": "Goa, India"
}

garbage_stats = []

def get_garbage_area(place):
    try:
        gdf_place = ox.geocode_to_gdf(place)

        if gdf_place is None or gdf_place.empty:
            raise ValueError("Geocoding failed or returned no result.")

        geometry = gdf_place.iloc[0].geometry

        if geometry is None or geometry.geom_type not in ['Polygon', 'MultiPolygon']:
            raise ValueError("Returned geometry is not a valid (Multi)Polygon.")

        polygon = geometry

        tags = {
            "landuse": ["landfill"],
            "amenity": ["waste_basket", "waste_disposal", "recycling"],
            "man_made": ["wastewater_plant"]
        }

        gdf = ox.features_from_polygon(polygon, tags)

        if gdf.empty:
            raise ValueError("No matching features. Check query location, tags, and log.")

        gdf_proj = gdf.to_crs(epsg=3857)
        garbage_area = gdf_proj.geometry.area.sum() / 10**6

        landfills = len(gdf[gdf['landuse'] == 'landfill'])
        disposals = len(gdf[gdf['amenity'] == 'waste_disposal'])
        recyclings = len(gdf[gdf['amenity'] == 'recycling'])
        baskets = len(gdf[gdf['amenity'] == 'waste_basket'])
        wastewater_plants = len(gdf[gdf['man_made'] == 'wastewater_plant'])

        gdf_place_proj = gdf_place.to_crs(epsg=3857)
        total_area = gdf_place_proj.geometry.area.iloc[0] / 10**6

        gdf_place = gdf_place.to_crs(epsg=4326)
        centroid = gdf_place.geometry.centroid.iloc[0]

        gdf_proj.to_crs(epsg=4326).plot(figsize=(6, 6), color='red', alpha=0.5)
        plt.title(f"Garbage Zones in {place}")
        plt.axis('off')
        plt.savefig(f"garbage_map_{place.replace(',', '').replace(' ', '_')}.png", bbox_inches='tight')
        plt.close()

        return round(garbage_area, 2), round(total_area, 2), centroid.y, centroid.x, landfills, disposals, recyclings, baskets, wastewater_plants

    except Exception as e:
        print(f"❌ Error in {place}: {e}")
        return 0.0, 0.0, None, None, 0, 0, 0, 0, 0

# Run and collect all locality data
def run_collection():
    for loc in localities:
        print(f"🗑️ Processing {loc}...")
        data = get_garbage_area(loc)
        if data[0] == 0.0 and loc in fallbacks:
            fallback = fallbacks[loc]
            print(f"🔁 Trying fallback for {loc} using {fallback}...")
            data = get_garbage_area(fallback)
        garbage_stats.append([loc] + list(data))
        print(f"✅ {loc}: {data[0]} km² garbage zone")

    columns = [
        "locality", "garbage_area_km2", "total_area_km2",
        "lat", "lon", "landfills", "disposals", "recyclings",
        "baskets", "wastewater_plants"
    ]
    df = pd.DataFrame(garbage_stats, columns=columns)

    df["garbage_ratio"] = df["garbage_area_km2"] / df["total_area_km2"].replace(0, 1)
    df.fillna(0, inplace=True)

    return df

# Visualize top garbage-heavy areas
def visualize_garbage(df):
    top_garbage = df.sort_values("garbage_ratio", ascending=False).head(10)
    plt.figure(figsize=(10, 6))
    plt.barh(top_garbage["locality"], top_garbage["garbage_ratio"], color="orange")
    plt.xlabel("Garbage Area Ratio")
    plt.title("Top 10 Garbage-heavy Areas")
    plt.gca().invert_yaxis()
    plt.show()

# Train and save ML model
def train_and_save_model(df):
    features = ["landfills", "disposals", "recyclings", "baskets", "wastewater_plants"]
    target = "garbage_ratio"

    X = df[features]
    y = df[target]

    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"✅ Model trained. MSE: {mse:.4f}")

    with open("greenai_garbage_model.pkl", "wb") as f:
        pickle.dump(model, f)
    print("✅ Model saved as greenai_garbage_model.pkl")

# Display map images inline
def show_map_images(df):
    print("\n📍 Garbage Zone Maps:\n")
    for locality in df["locality"]:
        image_name = f"garbage_map_{locality.replace(',', '').replace(' ', '_')}.png"
        image_path = os.path.join(".", image_name)
        if os.path.exists(image_path):
            print(f"📷 {locality}")
            display(Image(filename=image_path))

# Run the full pipeline
df = run_collection()
visualize_garbage(df)
train_and_save_model(df)
show_map_images(df)
```
